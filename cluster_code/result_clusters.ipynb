```python
from load_files_from_cluster import load_cluster_results
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update(
    {
        "font.size": 10,
        "text.usetex": True,
        "font.family": "serif",
        "font.serif": ["Times"],
    }
)
import cmasher as csh
aps_figwidth = 6.6
aps_halfwidth = 3.375
```

# 2D

```python
L1 = [20, 30, 70]
D = 2
kappa_shift = '0'
output_v0 = {}
for L_val in L1:
    output_v0[L_val] = load_cluster_results(
        f'results_{D}d_v_{kappa_shift}',
        f'results_sigma*_num_reals_100_L{L_val}.h5',
    )
kappa_shift = '3r0'
output_v3 = {}
L = [20, 30]
for L_val in L:
    output_v3[L_val] = load_cluster_results(
        f'results_{D}d_v_{kappa_shift}',
        f'results_sigma*_num_reals_100_L{L_val}.h5',
    )
```

```python
z2_v0 = {L_val: [np.mean(elem) for elem in output_v0[L_val]['z2']] for L_val in L1}
z2_v3 = {L_val: [np.mean(elem) for elem in output_v3[L_val]['z2']] for L_val in L}
```

```python

fig, ax = plt.subplots(
    1, 2,
    figsize=(1.2*aps_halfwidth, 0.4*aps_halfwidth),
    sharey=True,
    dpi=350,
    )



cmap = csh.bubblegum
for L_val in L1:
    ax[0].plot(
        output_v0[L_val]['sigma_values']*L_val,
        z2_v0[L_val],
        label=f'L={L_val}',
        marker='o', markersize=2, linewidth=.5,
        color=cmap((L_val - min(L1)) / (max(L1) - min(L1))),
        rasterized=True,
        )
for L_val in L:    
    ax[1].plot(
        output_v3[L_val]['sigma_values']*L_val,
        z2_v3[L_val],
        label=f'L={L_val}',
        marker='o', markersize=2, linewidth=.5,
        color=cmap((L_val - min(L)) / (max(L) - min(L))),
        rasterized=True,
        ) # {L_val**D} sites',

ax[0].set_ylabel(r'$Z_2$')
ax[0].set_xlabel(r'$\sigma (v=0)$')
ax[1].set_xlabel(r'$\sigma (v=3r_0)$')
# ax[1].set_ylabel(r'$Z_2$')
ax[0].legend(fontsize=9)
ax[0].set_yticks([-1, 0, 1])
ax[1].set_yticks([])
ax[0].set_xlim(0, 0.1)
ax[0].set_xticks([0, 0.05])
ax[1].set_xlim(0, 0.1)
ax[1].set_xticks([0, 0.05, 0.1])
# subplot adjusts:
fig.subplots_adjust(wspace=0.)
```

```python
fig.savefig("apx_finitesize_2D.png", bbox_inches='tight', dpi=350)
```
