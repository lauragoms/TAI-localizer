```python
from load_files_from_cluster import load_cluster_results
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update(
    {
        "font.size": 10,
        "text.usetex": True,
        "font.family": "serif",
        "font.serif": ["Times"],
    }
)
import cmasher as csh
aps_figwidth = 6.6
aps_halfwidth = 3.375
```

```python
L = [70]
D = 2
output = {}
for L_val in L:
    output[L_val] = load_cluster_results(
        f'results_{D}d_v_0',
        f'results_sigma*_num_reals_100_L{L_val}.h5',
    )
```

```python
z2 = {L_val: [np.mean(elem) for elem in output[L_val]['z2']] for L_val in L}
```

```python

import numpy as np

b=np.linspace(0, .1, 50)

E = len(b)
# for elem in b:
#     if elem not in a:
#         b.remove(elem)
# # print(b)
W = []
for i,elem in enumerate(b):
    # print(elem)
    if np.mod(i+1,E)==0:
        print('\n')
        lista = list(b[(i+1)-E:i+1])
        print(" ".join(f'{x}' for x in lista) )
        a = " ".join(f'"{[x]}"' for x in lista)
        W.append(a)
```

```python

plt.figure(figsize=(aps_halfwidth, aps_halfwidth), dpi=200)
# take colors from mAGMA COLORMAp:
cmap = csh.bubblegum
for L_val in L:
    plt.plot(output[L_val]['sigma_values']*L_val, z2[L_val], label=f'L={L_val}, {L_val**D} sites', marker='o', markersize=4, linewidth=1, color=cmap((L_val - min(L)) / (max(L) - min(L))))
plt.xlabel(r'$\sigma$')
plt.ylabel(r'Average $z_2$')
plt.title('System 3D, M=2, v=0')
plt.legend()
```
